```python
import pandas as pd
import numpy as np
import ast

df = pd.read_csv('multimodaltransformer_metrics.csv')

p = []
for i in df['#Compounds that have at least one match']:
    x = ast.literal_eval(i)
    p.append(x)
p = np.array(p)
print("Mean:", np.mean(p, axis=0))
print("Std:", np.std(p, axis=0))
```
