```python
import pandas as pd   
```

```python
seeds = [42, 123, 47899, 2025, 1, 20, 99, 1020, 345, 78] 

for seed in seeds:
    ATC_dataset = pd.read_csv(f'Rep_train_set{seed}.csv')
    ATC_dataset = ATC_dataset.drop(columns = ["ATC_level1", "ATC_level2", "ATC_level3", "ATC_level4"])
    column_ATC = ATC_dataset['ATC Codes']
    ATC_codes = column_ATC.copy()
    for index, codes in enumerate(column_ATC):
        splitted_code = codes.split('; ')
        ATC_codes.iloc[index] = splitted_code
    ATC_levels = pd.DataFrame(columns = ["ATC_level1", "ATC_level2", "ATC_level3", "ATC_level4"])
    code_level1 = set()
    code_level2 = set()
    code_level3 = set()
    code_level4 = set()
    
    for codes in ATC_codes:
        for code in codes:
            code_level1.add(code[0])
            code_level2.add(code[:3])
            code_level3.add(code[:4])
            code_level4.add(code[:5])
        ATC_levels.loc[len(ATC_levels)] = (list(code_level1), list(code_level2), list(code_level3), list(code_level4))
        ATC_levels = ATC_levels.reset_index(drop=True)
        code_level1 = set()
        code_level2 = set()
        code_level3 = set()
        code_level4 = set()
    
    ATC_code = pd.concat([ATC_dataset, ATC_levels], axis=1)
    
    # Save the updated DataFrame to a new file
    ATC_code.to_csv(f"SplitATC_Rep_train_set{seed}.csv", index=False)
```
